1. La altura de este animal es *no* menos de 1 pie, por lo que debe pasar por la siguiente etapa antes de que se le asigne una clase.
2. La altura de este animal es *no* menos de 2 pies, lo que significa que la condición se evalúa como "Falso" (es decir, "NO"). Por lo tanto, este animal es "LARGE".

Puede codificar fácilmente la primera condición como:

```python
# Primer condicional
if animal["height"] < 1: # Es mejor que 1 ft?
    print("SMALL")
else:
    print("Pasa a la siguiente etapa, este animal NO ES SMALL")
```

Esto funciona perfecto. Si desea que el código sea un poco más fácil de leer, puede reemplazar el `else ... if` por un atajo útil, **`elif`**. Como sugiere el nombre, `elif`  se puede usar cuando tiene un `if` inmediatamente después de un `else`. Entonces:

```python
# Primer condicional
if animal["height"] < 1:
  print("SMALL")
# Segundo condicional
elif animal["height"] <2:
  print("MEDIUM")
else:
  print("LARGE")
```

Este bloque tiene dos líneas de código menos. La lógica con este razonamiento:

1. Si la altura del animal es inferior a 1 pie, entonces es `SMALL`.
2. Si el animal no es `SMALL`, verifique si mide menos de 2 pies. Si es así, etiquételo como `MEDIUM`.
3. Si la altura del animal es diferente (es decir, no es `SMALL` ni `MEDIUM`), entonces tiene que ser `LARGE`.

## Comentarios sobre lógica booleana
---

Hasta ahora, nuestros condicionales han evaluado una sola condición para determinar qué líneas de código se ejecutan realmente cuando lo ejecutamos. Python también puede evaluar múltiples condiciones a la vez, usando lo que llamamos **[lógica booleana](https://www.pbs.org/video/oolean-logic-logic-gates-crash-course-computer-science-nobmpt/)**.

Hay cuatro operaciones booleanas que serán suficientes en la gran mayoría de los casos:

* $y$, que en Python es `and`,
* $o$, que en Python es `or`,
* **o exclusivo**, que en Python es `^`, y
* $no$, que en Python es `not`.

Esto está relacionado con las operaciones de conjuntos en una pasa lectura. Así, puede representar la lógica booleana [usando conjuntos](https://en.wikipedia.org/wiki/Boolean_algebra#Diagrammatic_representations).


### $AND$

Tienes dos variables booleanas `A` y` B`. Esta es la **tabla de verdad** relacionada con la operación $AND$ cuando se les aplica (una tabla de verdad es solo una matriz que muestra el resultado de una operación booleana para diferentes valores de `A` y `B`):

| Value of `A`	| Value of `B`  | Value of `A and B`|
|------------	|------------	|------------------	|
| False      	| False      	| False            	|
| False      	| True       	| False            	|
| True       	| False      	| False            	|
| True       	| True       	| True             	|

En otras palabras, `A and B` es `True` solo cuando ambas condiciones son `True`.

### $OR$

Esta es la tabla de verdad para la operación $OR$:

| Value of `A`	| Value of `B`  | Value of `A or B` |
|------------	|------------	|------------------	|
| False      	| False      	| False            	|
| False      	| True       	| True            	|
| True       	| False      	| True            	|
| True       	| True       	| True             	|


En otras palabras, `A o B` es siempre `True`, excepto cuando ambas condiciones son `False`.

### $XOR$

Esta es la tabla de verdad para la operación exclusiva o:

| Value of `A`	| Value of `B`  | Value of `A ^ B`  |
|------------	|------------	|------------------	|
| False      	| False      	| False            	|
| False      	| True       	| True            	|
| True       	| False      	| True            	|
| True       	| True       	| False            	|

En otras palabras, `A ^ B` es `True` solo cuando ambas condiciones tienen valores de verdad diferentes.

### $NOT$

La negación es el más fácil de los operadores. Simplemente invierte el valor de su booleano. Entonces, si `A` es `True`, entonces `not A` será `False`. Solo se necesita un argumento, por lo que no puede hacer `A not B` (pero *puede* hacer `not (A AND B) `, por ejemplo).
