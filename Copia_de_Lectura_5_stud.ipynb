## Funciones
---

¿Observó que tuvo que reescribir partes del código para que coincida con la nueva comida? Si tuviera que hacer esto para una Patilla, entonces tendría que modificar esas partes del código nuevamente. ¡Imagínese tener que hacer eso para todo el supermercado!

Las **funciones** pueden resultar útiles. Cuando tienes un código que te encuentras que es de uso recurrente, con solo pequeños ajustes, es mejor que escribas una función. Las funciones son excelentes si desea generalizar un código que pueda usarse en una variedad de situaciones. Por ejemplo, podemos convertir este código:

~~~python
if banana["calories"] < 5:
    calories_label = "CALORIE FREE"
elif banana["calories"] < 40:
    calories_label = "LOW CALORIE"
else:
    calories_label = None

print(calories_label)
~~~

en una función, de modo que no tenga que cambiar `banana` por el nombre del alimento cada vez que lo ejecute. En su lugar, puede tener un marcador de posición, como este:

```python
def assign_calories_label(food):

  '''
  Objetivo de la función
  Input : Descripción
  Output: Salida.
  '''
  if food["calories"] < 5: # Cambiamos "banana" for "food"
    calories_label = "CALORIE FREE"
  elif food["calories"] < 40: # Cambiamos "banana" for "food"
    calories_label = "LOW CALORIE"
  else:
    calories_label = None
  print(calories_label)
```

Observe que gran parte del código es el mismo que antes, pero "banana" ha sido reemplazado por "food", y hay una nueva línea `def assign_calories_label(food):` que precede a todo lo demás. Esta línea se conoce como **declaración de la función**. Siempre comienza con la palabra clave restringida **`def`**, seguida del nombre que queríamos para la función (en este caso, `assign_calories_label`), luego una colección de **argumentos** entre paréntesis, y finalmente termina con dos puntos. En este caso, `food` es el único argumento.

Puede pensar en los argumentos como si fueran parte de una conversación. Por ejemplo, el nombre de la función `assign_calories_label` es el tema de conversación y el argumento` food` es el tema de discusión. Entonces, si estuviéramos hablando de asignar una etiqueta de calorías y pensó "bueno, ¿cuál es la etiqueta de una `banana`, entonces "banana" se convierte en el argumento que **pasó** a nuestra conversación (*pasa* argumentos a funciones).

Si ejecuta la celda de arriba, no verá nada impreso. Una función está destinada a generalizar el código, pero sin darle información sobre nuestra situación específica para la que queremos usarla, ¡no sabrá qué hacer por nosotros! Ese es el propósito de los argumentos (en este caso, `food`): son marcadores de posición que podemos reemplazar con información específica de nuestra situación actual para que la función pueda darnos una salida relevante.

```python
assign_calories_label(tomato)
```

Es importante tener en cuenta que el código dentro de la función *debe tener indentación* para que funcione correctamente.

Nuestra función imprime algo en la pantalla, pero no tiene **output**; es decir, no produce nada de valor para la computadora. (La computadora no considera que algo que se muestra en la pantalla sea una salida, aunque esa definición puede parecernos natural ya que somos criaturas visuales, ¡esta es una distinción muy importante para recordar!)

Para que nuestra función devuelva una salida, podemos usar la palabra clave `return`. Esta nueva versión de nuestra función no imprime inmediatamente la etiqueta, sino que la devuelve como un objeto de cadena (o `None` si no se aplica ninguna etiqueta):

```python
def assign_calories_label(food):
    """
    Asigne una etiqueta de calorías de acuerdo con las reglas de la FDA.

    Argumentos:

    food: Un diccionario de Python que tiene al menos una clave de "calorías".

    Outputs:

    calories_label: Un string, entre "CALORIE FREE", "LOW CALORIE" or None
    """
    if food["calories"] < 5:
        calories_label = "CALORIE FREE"
    elif food["calories"] < 40:
        calories_label = "LOW CALORIE"
    else:
        calories_label = None
    return calories_label
```

Construya la función `assign_fat_label()` para implementar el árbol de decisión.

```python
def assign_fat_label(food):
  '''
  Clasifica según las calorias del alimento
  Agumento: (food), es un diccionario que debe contener la key "fat"
  Input: Etiqueta, es un string entre "FAT FREE", "LOW FAT" o None.
  '''
  if food["fat"] < 0.5:
    fat_food_label = "FAT FREE"
  elif food["fat"] < 3:
    fat_food_label = "LOW FAT"
  else:
    fat_food_label = None
  return fat_food_label
```

Resulta que nuestro conjunto de datos tiene contenido de grasa *en miligramos*, no en gramos. Por tanto, 250 mg equivalen a 0,25 gramos.

## Agregar etiquetas de azúcar
---

Para el azúcar, nuestra tarea es más sencilla: si hay menos de 0,5 gramos de azúcar por porción, es `SUGAR FREE`. Si hay más de 0,5 gramos, el alimento no recibe ninguna etiqueta. En código, esto es:

```python
def assign_sugar_label(food):
    """
    Asigna una etiqueta de azúcar de acuerdo con las reglas de la FDA.

    Argumentos:
    food: Un diccionario de Python que tiene al menos una clave de "azúcar".

    Salidas:
    sugar_label: Una cadena, ya sea "SUGAR FREE" o ninguna.
    """
    if food["sugar"] < 0.5:
        sugar_label = "SUGAR FREE"
    else:
        sugar_label = None
    return sugar_label
```

## Funciones anónimas (lambda functions)
---

Por lo general, no queremos crear funciones que no usaremos con frecuencia, por lo que podemos usar **funciones anónimas** en su lugar. Estas son como funciones "`def`"ordinarias, solo que no es necesario que les dé un nombre (por eso son anónimas). Para decirle a **Python** que algún fragmento de código será una función anónima, reemplazamos `def` con la palabra clave **`lambda`** y usamos la siguiente sintaxis:

~~~python
lambda my_input: <do_something_with_my_input> # Puede tener más de 1 entrada, solo sepárelas con coma.
~~~
La versión anónima de `assign_fda_labels(food)` puede ser:

~~~python
lambda food: [assign_calories_label(food), assign_fat_label(food), assign_sugar_label(food)]
~~~

y para ejecutarlo, haríamos esto:

```python
(lambda food: [assign_calories_label(food), assign_fat_label(food), assign_sugar_label(food)])(banana)
```

```python
f = (lambda food: [assign_calories_label(food), assign_fat_label(food), assign_sugar_label(food)])
f(banana)
```

```python
#f(tomato)
```

Depende de usted si desea utilizar una función ordinaria o una función `lambda`. Utilice lo que sea más legible y tenga sentido dado su contexto específico.
